```python
def simpleCleaning(fileDerectory):
    # data=pd.read_csv(fileDerectory,skiprows=1)
    data=pd.read_csv(fileDerectory)
    data.drop_duplicates(inplace=True)
    data.fillna("no value",inplace=True)
    return data
```

```python
simpleCleaning("dataset_olympics.csv")
```

```python
from math import ceil, floor

def devide_the_data(csv_porcentage,json_porcenntage,db_porcentage,lien_fichier,nom_csv_stocker,nom_json_stocker,nom_db_crier):
    handered_porcentage=csv_porcentage+json_porcenntage+db_porcentage
    if handered_porcentage>100 or handered_porcentage<100:
        print("la somme des pourcentage ne doit pas deppasser ou etre moi de 100 !")
    elif(type(lien_fichier)== str):
        df=simpleCleaning(lien_fichier)
        df_shuffled=df.sample(frac=1).reset_index(drop=True)
        print("shuffled",df_shuffled.head())
        total_lignes=len(df)
        csv_lignes=ceil((total_lignes*csv_porcentage)/100)
        print("csv_lignes :",csv_lignes)
        json_lignes=ceil((total_lignes*json_porcenntage)/100)
        print("json_lignes :",json_lignes)
        db_lignes=ceil((total_lignes*db_porcentage)/100)
        print("db_lignes :",db_lignes)
        all_ligns=csv_lignes+db_lignes+csv_lignes
        rest=total_lignes-all_ligns
        if rest != 0:
            csv_lignes=csv_lignes+rest
        
        print(csv_lignes)
        
        df_csv=df_shuffled.iloc[0:csv_lignes].to_csv(f"{nom_csv_stocker}.csv",index=False,header=False)
        df_json=df_shuffled.iloc[csv_lignes:(csv_lignes+json_lignes)].to_json(f"{nom_json_stocker}.json")
        df_db=df_shuffled.iloc[(csv_lignes+json_lignes):len(df_shuffled)].to_sql(f"{nom_db_crier}",engine, if_exists='replace',index=False)
        print("done!")
    else:
        print("something is not right")
```
