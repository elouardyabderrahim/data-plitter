```python
def simpleCleaning(fileDerectory):
    data=pd.read_csv(fileDerectory)
    data.drop_duplicates(inplace=True)
    data.fillna("no value",inplace=True)
    return data
```

```python
simpleCleaning("dataset_olympics.csv")
```
